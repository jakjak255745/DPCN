```
class Graph:
    def __init__(self):
        self.verts=[]
        self.edges=[]
    
    def addVertex(self,vert):
        if vert in self.verts:
            self.verts.append(str(vert) + str(1))
        
    def addVerticesFromList(self,vertList):
        for vert in vertList:
           self.addVert(vert)
        

    def addEdge(self,fromVert, toVert, weight=None):
        if weight==None:
            if fromVert not in self.verts:
                self.addVertex(fromVert) 
            elif toVert not in self.verts:
                self.addVertex(toVert)
            self.edges.append((fromVert,toVert)) 
        else:
            if fromVert not in self.verts:
                self.addVertex(fromVert) 
            elif toVert not in self.verts:
                self.addVertex(toVert)
            self.edges.append((fromVert,toVert,weight))

    def addEdgesFromList(self,edgeList):
        for edge in edgeList:
            if len(edge)==2:
                self.addEdge(edge[0],edge[1])
            if len(edge)==3:
                self.addEdge(edge[0],edge[1],edge[2])

    def getVerttices(self):
        print(self.verts)

    def getEdges(self):
        print(self.edges)

    def getNeighbors(self,vertKey):
        for e in self.edges:
            if e[0]==vertKey:
                neighbors.append(e[1])
            elif e[1]==vertKey:
                neighbors.append(e[0])
        return neighbors
    
    def __contains__(self, vert):
        return vert in self.verts

    def saveGraph(self,):
        result="graph G { \n"
        for e in self.edges:
            result+=str(e[0])+" -- "+str(e[1]) + "\n"
        result+="}"
        with open(str(self.__name__)+'.txt', 'w') as f:
            f.write(result)

    def getShortestPaths(self,fromVert):
        dist={}
        prev={}
        Q=[]#unvisited
        for v in self.verts:
            dist[v] = float('inf') 
            prev[v] = None
            Q.append(v)
        dist[fromVert] = 0
        i=0
        while len(Q)!=0:
            
            temp = [i for i in Q if min(dist,key=dist.get)==i]
            print(temp)
            u = min(dist,key=dist.get) if min(dist,key=dist.get) in Q else None
            Q.remove(u)
            
        
            for v in self.getNeighbors(u):
                if v in Q:
                    alt = dist[u] + 1
                    
                    if alt < dist[v]:
                        dist[v] = alt
                        prev[v] = u

        return dist
```
